```python
import numpy as np
from sklearn import metrics

def purity_score(y_train, y_pred):
    
    contingency_matrix = metrics.cluster.contingency_matrix(y_train, y_pred)
    
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix) 
```

```python
purity_score(y_test,y_pred)
```

```python
#it means 92.98% data total number of object classified correctly
#Purity=TP/(TP+TN+FP+FN) 
#TP=true positive,TN=true negative,FP=Flase posituve,FN=False negative
```
